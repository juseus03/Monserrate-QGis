```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
filesPath=np.zeros((0,0))
for file in os.listdir("./perfiles_ml"):
    if file.endswith(".txt"):
        filesPath=np.append(filesPath,os.path.join("./perfiles_ml", file))
```

```python
def findMax(dist,height):
    comp=np.less(dist,1230)
    height_trim=np.trim_zeros(height*comp)
    return height_trim[-1]
```

```python
max_points=np.zeros((0,0))
for path in filesPath:
    data=np.loadtxt(path)
    dist=data[:,0]
    height=data[:,3]
    max_points=np.append(max_points,findMax(dist,height))
    plt.plot(dist,height)
```

```python
angles=np.array([10.27,12.84,15.37,17.84,20.25,22.59,24.86,2.36,5.01,7.65,0,-0.3,-10.78,-13.31,-15.78,-18.19,-20.52,-22.79,-24.98,-2.96,-5.60,-8.21])
plt.scatter(angles*(-1)+65,max_points)
plt.xlim([45,85])
```

```python
data=np.loadtxt('perfil_frente.txt')
dist=data[:,0]
dist_transf=(dist/0.009)*50+40
height=data[:,3]
plt.scatter(angles*(-1)+65,max_points)
plt.plot(dist_transf,height)
```
