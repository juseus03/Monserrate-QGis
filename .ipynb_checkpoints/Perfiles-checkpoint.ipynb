```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
from scipy.interpolate import interp1d
```

```python
#Reads the file path
filesPath=np.zeros((0,0))
for file in os.listdir("./perfiles_ml"):
    if file.endswith(".txt"):
        filesPath=np.append(filesPath,os.path.join("./perfiles_ml", file))
```

```python
#Reads the mountain profile from files and get the maximum point
max_points=np.zeros((0,0))
for path in filesPath:
    data=np.loadtxt(path)
    dist=data[:,0]
    height=data[:,3]
    max_points=np.append(max_points,height[-1])
    plt.plot(dist,height)
```

```python
#Creates angle array to plot with maximum point array, it must be sorted!
angles=np.array([0,10.01,15.04,19.99,5.17,-10.01,-15.04,-19.97,-5.06])
angles=angles*(-1)+65
sort_index=np.argsort(angles)
sorted_angles=np.sort(angles)
sorted_max_points=np.zeros(len(max_points))
act_index=0
for i in sort_index:
    sorted_max_points[act_index]=max_points[i]
    act_index+=1
```

```python
#Reads front profile and compare it with the maximum point profile
data=np.loadtxt('perfil_frente.txt')
dist=data[:,0]
dist_transf=(dist/722.689533241)*39.96+45.01
height=data[:,3]

x_inp=np.linspace(45.1,84.97,num=100)
f2 = interp1d(dist_transf, height, kind='cubic')

f3=interp1d(sorted_angles,sorted_max_points, kind='cubic')

plt.errorbar(dist_transf,height,yerr=10,xerr=0.55,color='b',fmt='o')
plt.errorbar(angles,max_points,yerr=10,xerr=0.55,color='g',fmt='o')

plt.plot(x_inp,f2(x_inp),label='Cubic Spline Real Data')
plt.plot(x_inp,f3(x_inp),label='Cubic Spline Max Profile')

plt.xlim([42.5,85])
plt.title('Monserrate Frontal Profile')
plt.xlabel(r'$\phi$ $[^\circ]$')
plt.ylabel('Height [m]')
plt.grid()
plt.legend(loc=4)
```
